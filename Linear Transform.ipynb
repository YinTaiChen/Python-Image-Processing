```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

class imageLoader:
    def __init__(self, image):
        self.img = Image.open(image)
        self.gray = self.img.convert(mode="L")
        self.array_3d = np.array(self.img)
        self.array_1d = np.array(self.gray)
    
    def show(self):
        %matplotlib inline
        plt.imshow(self.img)
        
    def linearTransform(self, weight=1, bias=0, plot=False):
        output = self.array_3d * weight + bias
        output[output>255] = 255
        output[output<0] = 0
        
        if plot==True:
            img = Image.fromarray(np.uint8(output), mode='RGB')
            plt.imshow(output)
        elif plot==False:
            return output
```

```python
image = imageLoader("Lenna.png")
```

```python
image.show()
```

```python
image.linearTransform(weight=1.5, plot=True)
```

```python
image.linearTransform(weight=0.5, plot=True)
```

```python
image.linearTransform(bias=128, plot=True)
```

```python
image.linearTransform(bias=64, plot=True)
```

```python
image.linearTransform(plot=False)
```
