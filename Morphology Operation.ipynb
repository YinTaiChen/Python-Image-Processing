```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

class imageLoader:
    def __init__(self, image):
        %matplotlib inline
        self.img = Image.open(image)
        self.gray = self.img.convert(mode="L")
        self.array_3d = np.array(self.img)
        self.array_1d = np.array(self.gray)
           
    def show(self):
        plt.imshow(self.img)
        
    def morphOp(self, operation="open", radius=1, iteration=1):
        input = self.array_1d
        structure = cv2.getStructuringElement(cv2.MORPH_RECT, (2*radius+1, 2*radius+1))
        
        operation_dict = {
            "open": cv2.MORPH_OPEN,
            "close": cv2.MORPH_CLOSE,
            "gradient": cv2.MORPH_GRADIENT,
            "tophat": cv2.MORPH_TOPHAT,
            "blackhat": cv2.MORPH_BLACKHAT
        }
        
        result = cv2.morphologyEx(input, operation_dict[operation], structure, iterations=iteration)
        
        plt.imshow(result, cmap='gray')
```

```python
image = imageLoader("Lenna.png")
```

```python
image.show()
```
