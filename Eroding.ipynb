```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

class imageLoader:
    def __init__(self, image):
        %matplotlib inline
        self.img = Image.open(image)
        self.gray = self.img.convert(mode="L")
        self.array_3d = np.array(self.img)
        self.array_1d = np.array(self.gray)
           
    def show(self):
        plt.imshow(self.img)
        
    def eroding(self, size=3, plot="erode"):
        input = self.array_1d
        structure = cv2.getStructuringElement(cv2.MORPH_RECT, (size, size))
        result = cv2.erode(input, structure)
        edges = input - result
        
        if plot=="erode":
            plt.imshow(result, cmap='gray')
        elif plot=="edge":
            plt.imshow(edges, cmap='gray')
        elif plot=="original":
            plt.imshow(input, cmap='gray')
```

```python
image = imageLoader("Lenna.png")
```

```python
image.show()
```

```python
image.eroding(size=5)
```

```python
image.eroding(size=11)
```
