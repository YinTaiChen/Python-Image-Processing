```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import cv2

class imageLoader:
    def __init__(self, image):
        %matplotlib inline
        self.img = Image.open(image)
        self.gray = self.img.convert(mode="L")
        self.array_3d = np.array(self.img)
        self.array_1d = np.array(self.gray)
           
    def show(self):
        plt.imshow(self.img)
        
    def dilating(self, radius=1):
        input = self.array_1d
        structure = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2*radius+1, 2*radius+1))
        dilated = cv2.dilate(input, structure)
        
        plt.imshow(dilated, cmap='gray')
```

```python
image = imageLoader("Lenna.png")
```

```python
image.show()
```

```python
image.dilating(radius=3)
```

```python
image.dilating(radius=5)
```
